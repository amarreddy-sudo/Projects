```python
import os
from pathlib import Path
from icrawler.builtin import BingImageCrawler
import shutil

BASE_DIR = "/content/drive/MyDrive/food_project_1"
RAW_DIR = f"{BASE_DIR}/raw_images"
TEMP_DIR = f"{BASE_DIR}/temp_downloads"

os.makedirs(RAW_DIR, exist_ok=True)
os.makedirs(TEMP_DIR, exist_ok=True)

KEYWORDS = [
"indian thali full meal plate top view",
"south indian meals plate rice sambar",
"north indian thali rice chapati dal",
"indian lunch plate rice chapati curd",
"south indian banana leaf meal overhead",
"indian dinner thali plate multiple dishes",
"restaurant indian thali full overhead",
"homemade indian thali plate top view",
"indian food plate rice dal curry chapati",
"thali plate multiple dishes top view",

"indian rice chapati curry plate",
"indian vegetarian thali meal",
"indian lunch plate rice dal",
"traditional indian thali food",
"indian dinner plate rice chapati curry",
"restaurant indian meal plate",
"homemade indian food thali",
"indian buffet thali plate",
"indian meal platter overhead",
"indian veg thali full plate"
]

IMAGES_PER_KEYWORD = 80

image_counter = len(os.listdir(RAW_DIR))

for i, kw in enumerate(KEYWORDS):

    print(f"\nDownloading: {kw}")

    keyword_folder = f"{TEMP_DIR}/kw_{i}"
    os.makedirs(keyword_folder, exist_ok=True)

    crawler = BingImageCrawler(
        feeder_threads=4,
        parser_threads=4,
        downloader_threads=8,
        storage={"root_dir": keyword_folder}
    )

    crawler.crawl(keyword=kw, max_num=IMAGES_PER_KEYWORD)

    # move images to main folder with unique names
    for img in os.listdir(keyword_folder):
        src = os.path.join(keyword_folder, img)

        if img.lower().endswith((".jpg",".jpeg",".png")):
            new_name = f"img_{image_counter}.jpg"
            dst = os.path.join(RAW_DIR, new_name)

            shutil.move(src, dst)
            image_counter += 1

print("\nDownload complete!")

total = len(os.listdir(RAW_DIR))
print(f"Total downloaded images: {total}")
print(f"Saved at: {RAW_DIR}")
```

```python
from pathlib import Path
import io, cv2, json, shutil, random
import numpy as np
from PIL import Image
from IPython.display import display, clear_output
import ipywidgets as widgets

# CONFIG — only edit this section

CLASS_NAMES = [
    "rice",      # 0
    "chapathi",  # 1
    "curry",     # 2
]

IMG_DIR    = "/content/drive/MyDrive/food_project_1/raw_images"
FLAT_IMG   = "/content/drive/MyDrive/food_project_1/flat1/images"
FLAT_LBL   = "/content/drive/MyDrive/food_project_1/flat1/labels"
PROGRESS_F = "/content/drive/MyDrive/food_project_1/progress.json"

# ── How many images to label per session ─────────────────────────
IMAGES_PER_SESSION = 100   # change this to label more at once

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

Path(FLAT_IMG).mkdir(parents=True, exist_ok=True)
Path(FLAT_LBL).mkdir(parents=True, exist_ok=True)

COLORS = [
    (52,168,83),(66,133,244),(251,188,4),
    (234,67,53),(155,29,153),(0,172,193),
]

# ── Progress helpers ──────────────────────────────────────────────
def _load_prog():
    if Path(PROGRESS_F).exists():
        with open(PROGRESS_F) as f:
            return json.load(f)
    return {"labeled": [], "skipped": []}

def _save_prog(p):
    with open(PROGRESS_F, "w") as f:
        json.dump(p, f, indent=2)

def _get_remaining():
    p    = _load_prog()
    done = set(p["labeled"] + p["skipped"])
    imgs = sorted([
        x for x in Path(IMG_DIR).glob("*")
        if x.suffix.lower() in {".jpg", ".jpeg", ".png"}
    ])
    return [x for x in imgs if x.name not in done], p

# ── Draw boxes on image ───────────────────────────────────────────
def _draw(img_bgr, boxes):
    img = img_bgr.copy()
    ih, iw = img.shape[:2]
    for cls, (xc, yc, bw, bh) in boxes.items():
        cid = CLASS_NAMES.index(cls)
        x1  = max(0,      int((xc - bw/2) * iw))
        y1  = max(0,      int((yc - bh/2) * ih))
        x2  = min(iw - 1, int((xc + bw/2) * iw))
        y2  = min(ih - 1, int((yc + bh/2) * ih))
        col = COLORS[cid % len(COLORS)]
        cv2.rectangle(img, (x1,y1), (x2,y2), col, 3)
        (tw, th), _ = cv2.getTextSize(cls, cv2.FONT_HERSHEY_SIMPLEX, 0.75, 2)
        cv2.rectangle(img, (x1, max(0,y1-th-8)), (x1+tw+6, y1), col, -1)
        cv2.putText(img, cls, (x1+3, max(th, y1-3)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.75, (255,255,255), 2)
    return img

def _to_widget(img_bgr, max_w=660):
    ih, iw = img_bgr.shape[:2]
    sc     = min(1.0, max_w / iw)
    img    = cv2.resize(img_bgr, (int(iw*sc), int(ih*sc)))
    buf    = io.BytesIO()
    Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)).save(
        buf, format="JPEG", quality=90)
    return widgets.Image(value=buf.getvalue(), format="jpeg",
                         layout=widgets.Layout(width=f"{int(iw*sc)}px"))

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# MAIN LABELING FUNCTION
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def start_labeling():
    all_remaining, progress = _get_remaining()

    total_done = len(progress["labeled"])
    total_skip = len(progress["skipped"])
    total_left = len(all_remaining)

    print("=" * 55)
    print(f"  Total images      : {total_done + total_skip + total_left}")
    print(f"  Already labeled   : {total_done}")
    print(f"  Skipped           : {total_skip}")
    print(f"  Remaining         : {total_left}")
    print(f"  This session      : {min(IMAGES_PER_SESSION, total_left)}")
    print("=" * 55)

    if not all_remaining:
        print("All images are labeled! Proceed to training.")
        return

    # ── Take only IMAGES_PER_SESSION images this run ──────────────
    session_imgs = all_remaining[:IMAGES_PER_SESSION]

    state = {
        "idx":        0,
        "session":    session_imgs,       # only this session's images
        "all_remain": all_remaining,      # full remaining list
        "progress":   progress,
        "boxes":      {},
        "img_bgr":    None,
        "img_path":   None,
    }

    # ── Widgets ───────────────────────────────────────────────────
    status  = widgets.HTML()
    img_out = widgets.Output()

    # Session progress bar
    session_bar = widgets.IntProgress(
        value=0, min=0, max=len(session_imgs),
        description="Session:",
        bar_style="info",
        layout=widgets.Layout(width="400px")
    )
    session_lbl = widgets.Label(
        value=f"0 / {len(session_imgs)} done this session"
    )

    # Checkboxes — one per class
    cbs = {
        c: widgets.Checkbox(
            value=False,
            description=f"{i}: {c}",
            layout=widgets.Layout(width="160px")
        )
        for i, c in enumerate(CLASS_NAMES)
    }
    cb_grid = widgets.GridBox(
        list(cbs.values()),
        layout=widgets.Layout(
            grid_template_columns=f"repeat({min(3, len(CLASS_NAMES))}, 170px)",
            border="1px solid #bbb",
            padding="8px"
        )
    )

    # Slider panels — one per class
    sliders = {}

    def _make_panel(cls):
        kw = dict(
        min=0.01,
        max=0.99,
        step=0.01,
        continuous_update=True,   # IMPORTANT
        style={"description_width": "75px"},
        layout=widgets.Layout(width="390px")

        )
        s_xc = widgets.FloatSlider(value=0.50, description="x center", **kw)
        s_yc = widgets.FloatSlider(value=0.50, description="y center", **kw)
        s_bw = widgets.FloatSlider(value=0.30, description="width",    **kw)
        s_bh = widgets.FloatSlider(value=0.30, description="height",   **kw)

        def _upd(_):
            state["boxes"][cls] = (
                s_xc.value, s_yc.value, s_bw.value, s_bh.value)
            _refresh_img()

        for s in [s_xc, s_yc, s_bw, s_bh]:
            s.observe(_upd, names="value")

        cid   = CLASS_NAMES.index(cls)
        color = ["#2e7d32","#1565c0","#e65100",
                 "#6a1b9a","#00838f","#ad1457"][cid % 6]
        panel = widgets.VBox(
            [widgets.HTML(
                f"<b style='color:{color};font-size:14px'>"
                f"{cls} box</b>"),
             s_xc, s_yc, s_bw, s_bh],
            layout=widgets.Layout(
                border=f"2px solid {color}",
                border_radius="6px",
                padding="8px",
                margin="4px 0",
                display="none"
            )
        )
        return panel, (s_xc, s_yc, s_bw, s_bh)

    for cls in CLASS_NAMES:
        panel, refs = _make_panel(cls)
        sliders[cls] = (panel, *refs)

    sliders_box = widgets.VBox([sliders[c][0] for c in CLASS_NAMES])

    # Buttons
    b_save = widgets.Button(
        description="✔ SAVE & NEXT",
        button_style="success",
        layout=widgets.Layout(width="170px", height="40px")
    )
    b_skip = widgets.Button(
        description="⏭ SKIP",
        button_style="warning",
        layout=widgets.Layout(width="120px", height="40px")
    )
    b_clear = widgets.Button(
        description="✖ CLEAR",
        button_style="danger",
        layout=widgets.Layout(width="120px", height="40px")
    )
    btns = widgets.HBox(
        [b_save, b_skip, b_clear],
        layout=widgets.Layout(margin="10px 0")
    )

    tip = widgets.HTML("""
    <div style="background:#e3f2fd;border-left:4px solid #1976d2;
                padding:8px 12px;border-radius:6px;font-size:13px;
                margin-bottom:8px">
    <b>How to label:</b>&nbsp;
    ① Tick the checkbox for each food visible &nbsp;|&nbsp;
    ② Move sliders to position the box &nbsp;|&nbsp;
    ③ Watch live preview &nbsp;|&nbsp;
    ④ Click <b>✔ SAVE</b>
    </div>""")

    # ── Helper functions ──────────────────────────────────────────
    def _refresh_img():
        ann = _draw(state["img_bgr"], state["boxes"])
        with img_out:
            clear_output(wait=True)
            display(_to_widget(ann))

    def _refresh_status():
        idx    = state["idx"]
        tot    = len(state["session"])
        n_done = len(state["progress"]["labeled"])
        n_skip = len(state["progress"]["skipped"])
        name   = state["img_path"].name if state["img_path"] else ""
        left   = len(state["all_remain"]) - idx

        status.value = (
            f"<div style='background:#f1f8e9;padding:8px 12px;"
            f"border-radius:6px;font-size:13px;margin-bottom:6px'>"
            f"<b>Session image {idx+1} of {tot}</b>"
            f" &nbsp;│&nbsp; {name}"
            f" &nbsp;│&nbsp; Total labeled: {n_done}"
            f" &nbsp;│&nbsp; Remaining overall: {left}"
            f"</div>"
        )
        session_bar.value = idx
        session_lbl.value = f"{idx} / {tot} done this session"

    def _load(idx):
        # Session finished
        if idx >= len(state["session"]):
            remaining_after = len(state["all_remain"]) - len(state["session"])
            status.value = (
                f"<div style='background:#c8e6c9;padding:12px;"
                f"border-radius:6px;font-size:15px'>"
                f"<b>Session complete!</b> &nbsp;"
                f"Labeled {len(state['progress']['labeled'])} total. &nbsp;"
                f"{remaining_after} images still remaining.<br>"
                f"Run <code>start_labeling()</code> again to continue."
                f"</div>"
            )
            session_bar.value = len(state["session"])
            session_lbl.value = (
                f"{len(state['session'])} / {len(state['session'])} done ✓"
            )
            with img_out:
                clear_output()
            return

        state["idx"]      = idx
        state["img_path"] = state["session"][idx]
        state["img_bgr"]  = cv2.imread(str(state["img_path"]))
        state["boxes"]    = {}

        # Reset all checkboxes and hide slider panels
        for cls in CLASS_NAMES:
            cbs[cls].value                = False
            sliders[cls][0].layout.display = "none"

        _refresh_status()
        _refresh_img()

    # ── Checkbox handler ──────────────────────────────────────────
    def _on_cb(cls, change):
        panel = sliders[cls][0]
        if change["new"]:
            panel.layout.display = "flex"
            if cls not in state["boxes"]:
                _, s_xc, s_yc, s_bw, s_bh = sliders[cls]
                s_xc.value = 0.50
                s_yc.value = 0.50
                s_bw.value = 0.30
                s_bh.value = 0.30
                state["boxes"][cls] = (0.50, 0.50, 0.30, 0.30)
        else:
            panel.layout.display = "none"
            state["boxes"].pop(cls, None)
        _refresh_img()

    for cls in CLASS_NAMES:
        cbs[cls].observe(
            lambda ch, c=cls: _on_cb(c, ch), names="value")

    # ── Button handlers ───────────────────────────────────────────
    def _save(_):
        if not state["boxes"]:
            status.value = (
                "<div style='background:#fff3e0;padding:8px;"
                "border-radius:6px;color:#e65100'>"
                "Tick at least one food checkbox first!</div>"
            )
            return

        img_path = state["img_path"]
        stem     = f"meal_{img_path.stem}"

        # Save image copy
        shutil.copy2(img_path,
                     Path(FLAT_IMG) / (stem + img_path.suffix))

        # Save YOLO label file
        with open(Path(FLAT_LBL) / (stem + ".txt"), "w") as f:
            for cls, (xc, yc, bw, bh) in state["boxes"].items():
                cid = CLASS_NAMES.index(cls)
                f.write(f"{cid} {xc:.6f} {yc:.6f} {bw:.6f} {bh:.6f}\n")

        state["progress"]["labeled"].append(img_path.name)
        _save_prog(state["progress"])

        foods = list(state["boxes"].keys())
        print(f"✓ Saved [{state['idx']+1}/{len(state['session'])}]: "
              f"{stem} → {foods}")

        _load(state["idx"] + 1)

    def _skip(_):
        img_path = state["img_path"]
        state["progress"]["skipped"].append(img_path.name)
        _save_prog(state["progress"])
        print(f"⏭ Skipped [{state['idx']+1}/{len(state['session'])}]: "
              f"{img_path.name}")
        _load(state["idx"] + 1)

    def _clear(_):
        for cls in CLASS_NAMES:
            cbs[cls].value                = False
            sliders[cls][0].layout.display = "none"
        state["boxes"].clear()
        _refresh_img()

    b_save.on_click(_save)
    b_skip.on_click(_skip)
    b_clear.on_click(_clear)

    # ── Assemble UI ───────────────────────────────────────────────
    left_panel = widgets.VBox([
        widgets.HTML("<b>① Tick every food visible in the image:</b>"),
        cb_grid,
        widgets.HTML("<br><b>② Drag sliders to position each box:</b>"),
        sliders_box,
        btns,
        widgets.HBox([session_bar, session_lbl])
    ], layout=widgets.Layout(width="47%", padding="0 12px 0 0"))

    right_panel = widgets.VBox([
        widgets.HTML("<b>Live preview (updates as you move sliders):</b>"),
        img_out,
    ], layout=widgets.Layout(width="53%"))

    display(widgets.VBox([
        tip,
        status,
        widgets.HBox([left_panel, right_panel])
    ]))

    # Load first image
    _load(0)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# HELPER — check how many images are labeled so far
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
def labeling_status():
    p    = _load_prog()
    rem, _ = _get_remaining()
    print(f"Labeled  : {len(p['labeled'])}")
    print(f"Skipped  : {len(p['skipped'])}")
    print(f"Remaining: {len(rem)}")
    print(f"\nLabel files in flat/labels: "
          f"{len(list(Path(FLAT_LBL).glob('*.txt')))}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# RUN — call this to start
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
start_labeling()

# After each session, run this to check progress:
# labeling_status()
```

```python
# @title
import os, sys, shutil, yaml, cv2
import numpy as np
from pathlib import Path
from sklearn.model_selection import train_test_split
from ultralytics import YOLO
import torch

# ── Paths (same as Cell 3) ────────────────────────────────────────
BASE_DIR  = "/content/drive/MyDrive/food_project_1"
FLAT_IMG  = f"{BASE_DIR}/train/images"
FLAT_LBL  = f"{BASE_DIR}/train/labels"
SPLIT_DIR = f"{BASE_DIR}/dataset_split"
RUNS_DIR  = f"{BASE_DIR}/runs"

CLASS_NAMES = [
    "rice","chapati","curry"
]

# ════════════════════════════════════════════════════════════════════
# STEP 1: DIAGNOSE — print exactly what files exist and what matches
# ════════════════════════════════════════════════════════════════════

print("=" * 60)
print("  DIAGNOSIS")
print("=" * 60)

all_imgs = sorted([p for p in Path(FLAT_IMG).glob("*")
                   if p.suffix.lower() in {".jpg",".jpeg",".png"}])
all_lbls = sorted([p for p in Path(FLAT_LBL).glob("*.txt")])

print(f"  Images in flat/images : {len(all_imgs)}")
print(f"  Labels in flat/labels : {len(all_lbls)}")

# Build stem sets
img_stems = {p.stem: p for p in all_imgs}
lbl_stems = {p.stem: p for p in all_lbls}

matched   = [s for s in img_stems if s in lbl_stems]
img_only  = [s for s in img_stems if s not in lbl_stems]
lbl_only  = [s for s in lbl_stems if s not in img_stems]

print(f"\n  Matched pairs         : {len(matched)}")
print(f"  Images WITHOUT label  : {len(img_only)}")
print(f"  Labels WITHOUT image  : {len(lbl_only)}")

if img_only:
    print(f"\n  Sample unmatched images (first 5):")
    for s in img_only[:5]:
        print(f"    image stem: '{s}'")

if lbl_only:
    print(f"\n  Sample unmatched labels (first 5):")
    for s in lbl_only[:5]:
        print(f"    label stem: '{s}'")

# ════════════════════════════════════════════════════════════════════
# STEP 2: AUTO-REPAIR mismatches
# ════════════════════════════════════════════════════════════════════
# Common cause: image saved as meal_000001.jpg
#               label saved as meal_000001 (no suffix difference)
# OR image stem has different prefix than label stem
# We try fuzzy matching by stripping known prefixes and comparing base ids

print("\n" + "=" * 60)
print("  AUTO-REPAIR")
print("=" * 60)

repaired = 0

if img_only and lbl_only:
    # Build base-id maps (strip common prefixes like meal_, fullmeal_)
    def _base(stem):
        for prefix in ["meal_","fullmeal_","scene_","food_",""]:
            if stem.startswith(prefix):
                return stem[len(prefix):]
        return stem

    lbl_by_base = {_base(s): p for s,p in lbl_stems.items() if s in lbl_only}

    for img_stem in img_only:
        base = _base(img_stem)
        if base in lbl_by_base:
            # Found a match — rename label file to match image stem
            old_lbl = lbl_by_base[base]
            new_lbl = Path(FLAT_LBL) / (img_stem + ".txt")
            shutil.copy2(old_lbl, new_lbl)
            print(f"  Repaired: {old_lbl.name} → {new_lbl.name}")
            repaired += 1

print(f"  Repaired {repaired} label files.")

# ════════════════════════════════════════════════════════════════════
# STEP 3: Validate label file contents
# ════════════════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("  VALIDATING LABEL CONTENTS")
print("=" * 60)

# Rebuild after repair
all_imgs  = sorted([p for p in Path(FLAT_IMG).glob("*")
                    if p.suffix.lower() in {".jpg",".jpeg",".png"}])
img_stems = {p.stem: p for p in all_imgs}
lbl_stems = {p.stem: p for p in Path(FLAT_LBL).glob("*.txt")}
matched   = [s for s in img_stems if s in lbl_stems]

print(f"  Matched pairs after repair: {len(matched)}")

good   = []   # stems with valid labels
bad    = []   # stems with empty or corrupt labels
empty  = []   # label files with 0 lines

for stem in matched:
    lbl_path = lbl_stems[stem]
    with open(lbl_path) as f:
        lines = [l.strip() for l in f if l.strip()]

    if not lines:
        empty.append(stem)
        continue

    ok = True
    for line in lines:
        parts = line.split()
        if len(parts) != 5:
            bad.append(stem); ok = False; break
        try:
            cid = int(parts[0])
            vals = list(map(float, parts[1:]))
            if not (0 <= cid < len(CLASS_NAMES)):
                bad.append(stem); ok = False; break
            if not all(0.0 <= v <= 1.0 for v in vals):
                bad.append(stem); ok = False; break
        except:
            bad.append(stem); ok = False; break

    if ok:
        good.append(stem)

print(f"  Valid labels   : {len(good)}")
print(f"  Empty labels   : {len(empty)}")
print(f"  Corrupt labels : {len(bad)}")

if len(good) < 10:
    print("\n  ERROR: Less than 10 valid labeled images.")
    print("  Go back to Cell 4 and label more images.")
    print(f"  You need at least 30 images (currently {len(good)} valid).")
    raise SystemExit("Not enough labeled data to train.")

# ════════════════════════════════════════════════════════════════════
# STEP 4: Copy valid pairs to a CLEAN working directory
# ════════════════════════════════════════════════════════════════════
# This avoids YOLO silently skipping corrupt files during training

print("\n" + "=" * 60)
print("  PREPARING CLEAN DATASET")
print("=" * 60)

CLEAN_IMG = f"{BASE_DIR}/clean/images"
CLEAN_LBL = f"{BASE_DIR}/clean/labels"
os.makedirs(CLEAN_IMG, exist_ok=True)
os.makedirs(CLEAN_LBL, exist_ok=True)

# Clear old clean data
for f in Path(CLEAN_IMG).glob("*"): f.unlink()
for f in Path(CLEAN_LBL).glob("*"): f.unlink()

for stem in good:
    src_img = img_stems[stem]
    src_lbl = lbl_stems[stem]
    shutil.copy2(src_img, Path(CLEAN_IMG) / src_img.name)
    shutil.copy2(src_lbl, Path(CLEAN_LBL) / src_lbl.name)

print(f"  Copied {len(good)} clean pairs.")

# Class distribution
class_box_count = {c: 0 for c in CLASS_NAMES}
for stem in good:
    with open(lbl_stems[stem]) as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) == 5:
                cid = int(parts[0])
                if 0 <= cid < len(CLASS_NAMES):
                    class_box_count[CLASS_NAMES[cid]] += 1

print("\n  Bounding boxes per class:")
for cls, cnt in class_box_count.items():
    bar = "█" * min(cnt, 40)
    print(f"    {cls:<12}: {cnt:>4}  {bar}")

# ════════════════════════════════════════════════════════════════════
# STEP 5: Split into train / val / test
# ════════════════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("  SPLITTING DATASET  80 / 10 / 10")
print("=" * 60)

clean_imgs = sorted([p for p in Path(CLEAN_IMG).glob("*")
                     if p.suffix.lower() in {".jpg",".jpeg",".png"}])

# Shuffle before split
import random
random.seed(42)
random.shuffle(clean_imgs)

n     = len(clean_imgs)
n_val = max(1, int(n * 0.10))
n_tst = max(1, int(n * 0.10))
n_trn = n - n_val - n_tst

train_imgs = clean_imgs[:n_trn]
val_imgs   = clean_imgs[n_trn:n_trn+n_val]
test_imgs  = clean_imgs[n_trn+n_val:]

print(f"  Train : {len(train_imgs)}")
print(f"  Val   : {len(val_imgs)}")
print(f"  Test  : {len(test_imgs)}")

# Clear old split
if Path(SPLIT_DIR).exists():
    shutil.rmtree(SPLIT_DIR)

for split, imgs in [("train",train_imgs),("val",val_imgs),("test",test_imgs)]:
    id_ = Path(SPLIT_DIR)/split/"images"
    ld_ = Path(SPLIT_DIR)/split/"labels"
    id_.mkdir(parents=True, exist_ok=True)
    ld_.mkdir(parents=True, exist_ok=True)
    for p in imgs:
        shutil.copy2(p, id_/p.name)
        lsrc = Path(CLEAN_LBL)/(p.stem+".txt")
        if lsrc.exists():
            shutil.copy2(lsrc, ld_/lsrc.name)

# Verify split counts
for split in ["train","val","test"]:
    ni = len(list((Path(SPLIT_DIR)/split/"images").glob("*")))
    nl = len(list((Path(SPLIT_DIR)/split/"labels").glob("*.txt")))
    print(f"  {split:6s} → {ni} images, {nl} labels  {'✓' if ni==nl else '✗ MISMATCH'}")

# ════════════════════════════════════════════════════════════════════
# STEP 6: Write data.yaml
# ════════════════════════════════════════════════════════════════════

data_yaml_dict = {
    "path":  str(Path(SPLIT_DIR).resolve()),
    "train": "train/images",
    "val":   "val/images",
    "test":  "test/images",
    "nc":    len(CLASS_NAMES),
    "names": CLASS_NAMES,
}
yaml_path = f"{SPLIT_DIR}/data.yaml"
with open(yaml_path,"w") as f:
    yaml.dump(data_yaml_dict, f, default_flow_style=False, sort_keys=False)
print(f"\n  data.yaml saved → {yaml_path}")

# ════════════════════════════════════════════════════════════════════
# STEP 7: TRAIN
# ════════════════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("  TRAINING YOLOv8")
print("=" * 60)

device = 0 if torch.cuda.is_available() else "cpu"
print(f"  Device : {'GPU ✓' if device==0 else 'CPU (slow)'}")
print(f"  Images : {len(train_imgs)} train / {len(val_imgs)} val")

model = YOLO("yolov8n.pt")

model.train(
    data       = yaml_path,
    epochs     = 50,
    imgsz      = 640,
    batch      = 8,          # reduced from 16 — safer for small dataset
    project    = RUNS_DIR,
    name       = "food_model",
    exist_ok   = True,
    patience   = 20,
    device     = device,
    workers    = 0,          # required for Colab
    save       = True,
    plots      = True,
    # Augmentation — critical for multi-food detection
    hsv_h      = 0.015,
    hsv_s      = 0.7,
    hsv_v      = 0.4,
    fliplr     = 0.5,
    mosaic     = 1.0,        # combines 4 images → model sees multiple foods
    mixup      = 0.1,
    copy_paste = 0.3,        # pastes objects across images
)

# Find weights
BEST_PT = f"{RUNS_DIR}/food_model/weights/best.pt"
if not Path(BEST_PT).exists():
    candidates = sorted(Path(RUNS_DIR).glob("*/weights/best.pt"))
    if candidates:
        BEST_PT = str(candidates[-1])

print(f"\n  Best weights → {BEST_PT}")

# Validate
val_model = YOLO(BEST_PT)
metrics   = val_model.val(data=yaml_path, split="test", verbose=False)
print(f"\n  mAP50    : {metrics.box.map50:.4f}")
print(f"  mAP50-95 : {metrics.box.map:.4f}")
print(f"  Precision: {metrics.box.mp:.4f}")
print(f"  Recall   : {metrics.box.mr:.4f}")
print("\n  Done! Run Cell 6 to predict calories on a meal photo.")
```

```python
import cv2, io, os, numpy as np
from pathlib import Path
from PIL import Image
from IPython.display import display
from google.colab import files
from ultralytics import YOLO
# Base project folder in Drive
BASE_DIR = "/content/drive1/MyDrive/food_project"
BEST_PT = "/content/drive/MyDrive/food_project_1/runs/food_model/weights/best.pt"

# Where YOLO training saved runs
WEIGHTS_DIR = f"{BASE_DIR}/runs"
# ── Calorie database (ICMR values) ───────────────────────────────
CALORIE_DB = {
    "rice":    (206, "1 cup cooked ~180g"),
    "chapati": (104, "1 medium ~40g"),
    "curry":    (150, "1 cup ~200g")
}

COLORS = [
    (52,168,83),(66,133,244),(251,188,4),(234,67,53),(155,29,153),
    (0,172,193),(255,109,0),(46,204,113),(231,76,60),(52,73,94)
]

# ── Find best.pt automatically ───────────────────────────────────
BEST_PT = BEST_PT
if not Path(BEST_PT).exists():
    candidates = sorted(Path(WEIGHTS_DIR).glob("*/weights/best.pt"))
    if candidates:
        BEST_PT = str(candidates[-1])

if not Path(BEST_PT).exists():
    raise FileNotFoundError(
        "Trained weights not found.\n"
        "Complete Cell 5 (training) first."
    )

print(f"Model loaded from: {BEST_PT}")
model = YOLO(BEST_PT)
print(f"Classes: {list(model.names.values())}")

# ── Upload image ──────────────────────────────────────────────────
print("\nUpload your meal photo:")
uploaded = files.upload()

img_name = list(uploaded.keys())[0]
img_path = f"/content/{img_name}"
print(f"Uploaded: {img_name}")

# ── Detect ────────────────────────────────────────────────────────
print("Detecting food items...")

# Run at 2 scales and merge — catches items at different sizes
all_detections = []

for imgsz in [640, 1280]:
    res = model.predict(
        source  = img_path,
        conf    = 0.15,       # low threshold — per-class filter below
        iou     = 0.30,       # lower NMS — prevents boxes merging
        imgsz   = imgsz,
        verbose = False,
    )
    for box in res[0].boxes:
        cid  = int(box.cls[0])
        name = model.names[cid]
        conf = float(box.conf[0])
        coords = tuple(map(int, box.xyxy[0]))
        all_detections.append({"name":name,"conf":conf,"box":coords,"scale":imgsz})

# Per-class confidence thresholds
PER_CLASS_CONF = {
    "rice":   0.10,  # rice is uniform — needs lower threshold
    "chapathi": 0.12,
    "curry":0.15
}
DEFAULT_CONF = 0.25

# Filter + deduplicate across scales
def _iou(b1,b2):
    ix1=max(b1[0],b2[0]); iy1=max(b1[1],b2[1])
    ix2=min(b1[2],b2[2]); iy2=min(b1[3],b2[3])
    inter=max(0,ix2-ix1)*max(0,iy2-iy1)
    a1=(b1[2]-b1[0])*(b1[3]-b1[1]); a2=(b2[2]-b2[0])*(b2[3]-b2[1])
    return inter/(a1+a2-inter+1e-6)

kept = []
for d in all_detections:
    thr = PER_CLASS_CONF.get(d["name"], DEFAULT_CONF)
    if d["conf"] < thr:
        continue
    dup = False
    for k in kept:
        if k["name"]==d["name"] and _iou(k["box"],d["box"])>0.50:
            if d["conf"]>k["conf"]: kept.remove(k)
            else: dup=True
            break
    if not dup:
        kept.append(d)

# ── Count per class ───────────────────────────────────────────────
class_count = {}
for d in kept:
    class_count[d["name"]] = class_count.get(d["name"],0)+1

# ── Calculate calories ────────────────────────────────────────────
total_cal  = 0
rows       = []
for name, count in sorted(class_count.items()):
    cal_each, serving = CALORIE_DB.get(name,(0,"unknown"))
    item_total        = cal_each * count
    total_cal        += item_total
    rows.append((name, count, cal_each, serving, item_total))

# ── Print table ───────────────────────────────────────────────────
sep = "─" * 66
print(f"\n{sep}")
print(f"{'Food':<14} {'Count':>5}  {'Cal/serving':>11}  {'Serving':^20}  {'Total':>6}")
print(sep)

if not rows:
    print("  No food detected.")
    print(f"  DEBUG — raw detections before filter:")
    for d in all_detections:
        thr = PER_CLASS_CONF.get(d["name"],DEFAULT_CONF)
        st  = "KEPT" if d["conf"]>=thr else "DROPPED"
        print(f"    [{st}] {d['name']:<12} conf={d['conf']:.3f} thr={thr:.2f}")
else:
    for name,count,cal_each,serving,item_total in rows:
        print(f"{name:<14} {count:>5}  {cal_each:>9} cal  {serving:^20}  {item_total:>4} cal")

print(sep)
print(f"{'TOTAL CALORIES':>55}  {total_cal:>4} cal")
print(sep)

# ── Annotate image ────────────────────────────────────────────────
img = cv2.imread(img_path)
ih,iw = img.shape[:2]

for d in kept:
    cid        = list(model.names.values()).index(d["name"])
    x1,y1,x2,y2 = d["box"]
    x1,y1 = max(0,x1),max(0,y1)
    x2,y2 = min(iw-1,x2),min(ih-1,y2)
    cal    = CALORIE_DB.get(d["name"],(0,))[0]
    col    = COLORS[cid%len(COLORS)]
    label  = f"{d['name']}  {cal}cal  {d['conf']:.0%}"
    fs     = max(0.5, min(iw,ih)/1000)
    cv2.rectangle(img,(x1,y1),(x2,y2),col,2)
    (tw,th),_ = cv2.getTextSize(label,cv2.FONT_HERSHEY_SIMPLEX,fs,1)
    cv2.rectangle(img,(x1,max(0,y1-th-10)),(x1+tw+8,y1),col,-1)
    cv2.putText(img,label,(x1+4,max(th+2,y1-4)),
                cv2.FONT_HERSHEY_SIMPLEX,fs,(255,255,255),1)

# Total banner
banner    = np.zeros((50,iw,3),dtype=np.uint8)
banner[:] = (20,20,20)
cv2.putText(banner,f"  TOTAL CALORIES: {total_cal} cal",
            (10,34),cv2.FONT_HERSHEY_SIMPLEX,0.9,(50,255,150),2)
img_final = np.vstack([img,banner])

# Display inline
display(Image.fromarray(cv2.cvtColor(img_final,cv2.COLOR_BGR2RGB)))

# Save + download
stem     = Path(img_name).stem
out_path = f"/content/{stem}_result.jpg"
cv2.imwrite(out_path, img_final)

# Copy to Drive
try:
    shutil.copy2(out_path, f"{BASE_DIR}/{stem}_result.jpg")
    print(f"Result saved to Drive: {BASE_DIR}/{stem}_result.jpg")
except: pass

files.download(out_path)
print(f"\nDone! Total: {total_cal} calories")
```
